```python
# Import necessary libraries
import seaborn as sns                     # Seaborn for data visualization
import matplotlib.pyplot as plt           # Matplotlib for displaying plots
```

```python
# Load the built-in "tips" dataset from seaborn
tips = sns.load_dataset("tips")           # Contains data about restaurant bills and tips
```

```python
# Display the first few rows to understand the dataset
print(tips.head())                        # Shows first 5 rows (columns: total_bill, tip, sex, smoker, day, time, size)
```

```python
# ðŸŸ¢ Scatter plot - Total Bill vs Tip
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
# x-axis â†’ total bill amount
# y-axis â†’ tip amount
# hue="sex" â†’ colors the points based on gender (male/female)
plt.title("Total Bill vs Tip")            # Sets the plot title
plt.show()                                # Displays the scatter plot
```

```python
# ðŸŸ¡ Boxplot - Tip amount by day
sns.boxplot(x="day", y="tip", data=tips)
# x-axis â†’ Day of the week (Thur, Fri, Sat, Sun)
# y-axis â†’ Tip amount
# Each box shows median, quartiles, and outliers for tips given that day
plt.title("Tip Distribution by Day")
plt.show()
```

```python
# ðŸ”µ Bar plot - Average tip by smoker/non-smoker
sns.barplot(x="smoker", y="tip", data=tips, estimator="mean")
# x-axis â†’ 'smoker' (Yes or No)
# y-axis â†’ average tip (Seaborn automatically computes mean)
# estimator="mean" â†’ ensures it shows average values
plt.title("Average Tip by Smoker Status")
plt.show()
```
