```python
# Import necessary libraries
import seaborn as sns                     # Seaborn for data visualization
import matplotlib.pyplot as plt           # Matplotlib for displaying plots
```

```python
# Load the built-in "tips" dataset from seaborn
tips = sns.load_dataset("tips")           # Contains data about restaurant bills and tips
```

**Who is visiting mostly to restaurant male or female?**


```python
gender_counts=tips["sex"].value_counts()
tips_amount=tips["tip"]
plt.figure(figsize=(10,5))
plt.pie(gender_counts,labels=gender_counts.index,autopct="%1.1f%%")
plt.title("  Distribution of sex")
plt.show()
```

```python
# 🟢 Scatter plot - Total Bill vs Tip
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
# x-axis → total bill amount
# y-axis → tip amount
# hue="sex" → colors the points based on gender (male/female)
plt.title("Total Bill vs Tip")            # Sets the plot title
plt.show()                                # Displays the scatter plot
```

```python
# 🟡 Boxplot - Tip amount by day
sns.boxplot(x="day", y="tip", data=tips)
# x-axis → Day of the week (Thur, Fri, Sat, Sun)
# y-axis → Tip amount
# Each box shows median, quartiles, and outliers for tips given that day
plt.title("Tip Distribution by Day")
plt.show()
```

```python
# 🔵 Bar plot - Average tip by smoker/non-smoker
sns.barplot(x="smoker", y="tip", data=tips, estimator="mean")
# x-axis → 'smoker' (Yes or No)
# y-axis → average tip (Seaborn automatically computes mean)
# estimator="mean" → ensures it shows average values
plt.title("Average Tip by Smoker Status")
plt.show()
```

**On which time most tip is given**?

```python
tips.time.value_counts()
```

```python
sns.barplot(x="time",y="tip",data=tips,estimator="sum")
plt.title("Highest Tip by Time")
plt.show()
```

#                                        **Key Insights**

*   Males visite mostly to restaurant.
*   Males tend to give higher tip as compared to females.
*   While Saturday receives the highest total tip amount, Sunday sees the most indivisual tips.
*   Based on the bar plot, the average tip from smokers appears to be slightly higher than from non-smokers
*   Dinner time receives significantly more tips than lunch time.
